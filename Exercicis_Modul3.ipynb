```python
#Mostrem les 10 companyies amb la mitjana de retards més gran. 
df_sample.groupby('UniqueCarrier')['ArrDelay'].mean().sort_values(ascending=False).head(10)
```

```python
#Aquí obtenim la suma de minuts de retard per cada companyia
df_sample.groupby('UniqueCarrier')['ArrDelay'].sum().sort_values(ascending=False).head(10)
```

```python
#Aquí comptem el nombre de retards.
df_sample.groupby('UniqueCarrier')['ArrDelay'].count().sort_values(ascending=False).head(10)
```

Aquests càlculs ens permeten estudiar el concepte de retard amb diferents punts de vista. En funció del tipus d'encàrrec, es poden utilizar uns o altres resultats. El problema d'aquests càlculs és que tenen en compte retards inferiors a 15 minuts, els quals no els hem considerat retard, i seria convenient que quedessin fora de l'estudi.

```python
#Creem la columna Delayed, indicant si aquest retard és cert o fals, en funció de si és menor a 15 minuts o no.
df_sample['Delayed'] = df_sample['ArrDelay'] >= 15
```
